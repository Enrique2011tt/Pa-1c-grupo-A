```python
#Cálculo de desconto
valor_compra = float(input("Digite o valor da compra: "))


if valor_compra <= 100:
    desconto = 0
elif valor_compra <= 500:
    desconto = 0.10  
else:
    desconto = 0.20  

valor_final = valor_compra * (1 - desconto)

print(f"Desconto aplicado: {desconto*100:.0f}%")
print(f"Valor final a pagar: R$ {valor_final:.2f}")
```

```python
#Verificação de triângulo

lado1 = float(input("Digite o comprimento do primeiro lado do triângulo: "))
lado2 = float(input("Digite o comprimento do segundo lado do triângulo: "))
lado3 = float(input("Digite o comprimento do terceiro lado do triângulo: "))

if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
    print("Os lados formam um triângulo válido.")

    if lado1 == lado2 and lado2 == lado3:
        print("É um triângulo Equilátero (todos os lados iguais).")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("É um triângulo Isósceles (dois lados iguais).")
    else:
        print("É um triângulo Escaleno (todos os lados diferentes).")
else:
    print("Os lados não formam um triângulo válido. A soma de dois lados deve ser maior que o terceiro lado.")
```

```python
#Sistema de energia

energia = int(input("Digite o nível de energia (0-100): "))

if energia == 0:
    status = "Morto"
elif 1 <= energia <= 30:
    status = "Crítico"
elif 31 <= energia <= 70:
    status = "Normal"
elif 71 <= energia <= 100:
    status = "Saudável"

print(f"Status do personagem: {status}")
```

```python
#Média escolar


media = float(input("Digite a média final: "))

if media >= 7:
    situacao = "Aprovado"
elif 5 <= media < 7:
    situacao = "Recuperação"
else:
    situacao = "Reprovado"


print(f"Situação: {situacao}")
```

```python
#Cálculo de salário com bônus

salario = float(input("Digite o salário do funcionário: R$ "))
pontuacao = int(input("Digite a pontuação de desempenho (0-10) "))

if pontuacao < 5:
    percentual = 0
elif 5 <= pontuacao <= 7:
    percentual = 0.05  
elif 8 <= pontuacao <= 9:
    percentual = 0.10  
elif pontuacao == 10:
    percentual = 0.15  
else:
    percentual = -1 

if percentual != -1:
    bonus = salario * percentual
    salario_total = salario + bonus
    print(f"Bônus aplicado: {percentual*100:.0f}% (R$ {bonus:.2f})")
    print(f"Novo salário: R$ {salario_total:.2f}")
else:
    print("Erro: Pontuação fora do intervalo (0 a 10).")
```
