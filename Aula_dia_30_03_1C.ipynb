```python
#escrever um programa que classifica os triângulos

#verificar se forma um triângulo válido

#tipo:

#equilátero
#Isósceles
#escaleno

 #Entrada

lado1 = int (input("Digite o lado 1: "))
lado2 = int (input("Digite o lado 2: "))
lado3 = int (input("Digite o lado 3: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:

  print ("O triângulo é válido")

  if lado1 == lado2 and lado2 == lado3:

    print ("O triângulo é equilátero")

  elif lado1 == lado2 and lado2 != lado3:

    print ("O triângulo é isósceles")

  else:

    print ("O triângulo é escaleno")

else:
  print ("O triângulo não é válido")

```
