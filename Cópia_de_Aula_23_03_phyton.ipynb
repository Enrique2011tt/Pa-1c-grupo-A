```python
nota_01 = float(input("digite a nota do 01 : "))
nota_02 = float(input("digite a nota do 02 : "))
nota_03 = float(input("digite a nota do 03 : "))
nota_04 = float(input("digite a nota do 04 : "))

média = (nota_01 + nota_02 + nota_03 + nota_04) / 4

if média >= 7:
  print(f"aprovado com média {média}")
  print(f"parabéns!")
else:
  print(f"você está de recuperação {média}")
```
